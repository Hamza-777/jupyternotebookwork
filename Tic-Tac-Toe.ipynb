```python
# HERE GOES THE ACTUAL GAME

# Importing some useful libraries
from IPython.display import clear_output
import random

# Function to decide which player gets to play first
def decide_turn():
    x=random.randint(1,2)
    if x==1:
        return 'Player 1'
    else:
        return 'Player 2'

# Function to choose a character for the player who plays first    
def decide_character():
    while True:
        character=input("Choose your character ('X' or 'O') : ").upper()
        if character=='X' or character=='O':
            break
        elif character!='X' or character!='O':
            clear_output()
            print('Please enter a valid character!!!')
    return character

# Function to confirm if the players want to start playing
def start_game():
    while True:
        x=input('Start Game ? (YES or NO) :').upper()
        if x=='YES' or x=='NO':
            break
        else:
            clear_output()
            print("Enter Valid Input!!!")
    if x=='YES':
        return True
    else:
        return False

# Function to take the position where the players want to place their characters
def position(name,d):
    while True:
        x=input(f'Enter the position for {name}: ')
        if x.isdigit() and int(x) in range(1,10):
            if d[x]!='_':
                clear_output()
                layout(d)
                print("Position Already Occupied")
            else:
                break
        else:
            clear_output()
            layout(d)
            print("Enter Valid Position!!!")
    return x

# Function to display the game 
def layout(d):
    print("                                        |       |       ")
    print("                                    "+str(d['1'])+"   |   "+str(d['2'])+"   |   "+str(d['3'])+"   ")
    print('                                 ------------------------')
    print("                                        |       |       ")
    print("                                    "+str(d['4'])+"   |   "+str(d['5'])+"   |   "+str(d['6'])+"   ")
    print('                                 ------------------------')
    print("                                        |       |       ")
    print("                                    "+str(d['7'])+"   |   "+str(d['8'])+"   |   "+str(d['9'])+"   ")

# Function to decide the winner of the game or if the game resulted in a tie    
def win(d,player):
    if d['1']==d['2']==d['3']!='_' or d['4']==d['5']==d['6']!='_' or d['7']==d['8']==d['9']!='_' or d['1']==d['4']==d['7']!='_' or d['2']==d['5']==d['8']!='_' or d['3']==d['6']==d['9']!='_' or d['1']==d['5']==d['9']!='_' or d['3']==d['5']==d['7']!='_':
        print(f'{player} wins !')
        return True
    elif d['1']!='_' and d['2']!='_' and d['3']!='_' and d['4']!='_' and d['5']!='_' and d['6']!='_' and d['7']!='_' and d['8']!='_' and d['9']!='_':
        print("It's a tie!!!")
        return True
    else:
        return False

# Function to ask the players if they want to play again    
def playagain():
    while True:
        x=input('Want to play again ? (YES or NO) :').upper()
        if x=='YES' or x=='NO':
            break
        else:
            clear_output()
            print("Enter Valid Input!!!")
    if x=='YES':
        return True
    else:
        return False

# All the above functions are combined here to make the actual flow of game    
def tic_tac_toe():
    print("                                                    WELCOME TO TIC-TAC-TOE!!!!!!")
    turn1=decide_turn()
    if turn1=='Player 1':
        turn2='Player 2'
    else:
        turn2='Player 1'
    print(f'{turn1} goes first')
    chr1=decide_character()
    if chr1=='X':
        chr2='O'
    else:
        chr2='X'
    d={'1':'_','2':'_','3':'_','4':'_','5':'_','6':'_','7':'_','8':'_','9':'_'}   
    if start_game():
        while True:
            pos1=position(turn1,d)
            d[str(pos1)]=chr1
            clear_output()
            layout(d)
            if win(d,turn1):
                if not playagain():
                    break
                else:
                    tic_tac_toe()
            pos2=position(turn2,d)
            d[str(pos2)]=chr2
            clear_output()
            layout(d)
            if win(d,turn2):
                if not playagain():
                    break
                else:
                    tic_tac_toe()
        
tic_tac_toe()
```
