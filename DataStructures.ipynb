```python
# SINGLY lINKED LIST

class node:
    def __init__(self,data):
        self.data=data
        self.next=None

class linkedlist:
    def __init__(self):
        self.head=None
        
    def printlist(self):
        temp=self.head
        while(temp):
            print(temp.data)
            temp=temp.next
            
    def atBeginning(self,newData):
        newNode=node(newData)
        newNode.next=self.head
        self.head=newNode
    
    def atEnd(self,newData):
        newNode=node(newData)
        if self.head is None:
            self.head=newNode
            return
        temp=self.head
        while(temp.next):
            temp=temp.next
        temp.next=newNode  
    
    def atPosition(self,position,newData):
        newNode=node(newData)
        temp=self.head
        current=0
        while(current!=position-1):
            temp=temp.next
            current+=1
            if temp is None:
                print('No can do')
                return
        newNode.next=temp.next
        temp.next=newNode
        
    def remove(self,removalData):
        temp=self.head
        if temp is None:
            return
        if temp is not None:
            if temp.data==removalData:
                self.head=temp.next
                temp=None
                return
        while temp is not None:
            if temp.data==removalData:
                break
            prev=temp
            temp=temp.next
        
        if temp is None:
            return
        
        prev.next=temp.next
        temp=None
     
    def deleteAtBeg(self):
        temp=self.head
        if temp is None:
            return -1
        else:
            self.head=temp.next
            temp=None
            
    def deleteAtEnd(self):
        temp=self.head
        if temp is None:
            return -1
        else:
            while(temp.next):
                prev=temp
                temp=temp.next
            prev.next=None
            temp=None
    def deleteAtPosition(self,position):
        temp=self.head
        current=0
        if temp is None:
            return -1
        else:
            while(current!=position):
                prev=temp
                temp=temp.next
                current+=1
                if temp is None:
                    print('No element at required position')
                    return
            prev.next=temp.next
            temp=None
                
                
        


llist=linkedlist()
llist.head=node(1)
second=node(2)
third=node(3)
llist.head.next=second
second.next=third
llist.atBeginning(0)
llist.atEnd(4)
llist.deleteAtBeg()
llist.deleteAtEnd()
llist.deleteAtPosition(1)
llist.deleteAtPosition(2)
llist.printlist()
```

```python
# STACK
# can also be implemented using collections.deque which is an optimal way
class stack:
    def __init__(self):
        self.stack=[]
        
    def peek(self):
        return self.stack[-1]
    
    def pop(self):
        if len(self.stack)<=0:
            return -1
        else:
            return self.stack.pop()
    
    def push(self,number):
        self.stack.append(number)
            
s=stack()
s.push(1)
s.push(2)
s.peek()
s.pop()
s.peek()
```
